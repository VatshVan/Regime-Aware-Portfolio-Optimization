# NIFTY 50 OHLCV

```python
import pandas as pd
import yfinance as yf

# Current NIFTY 50 constituents
url = "https://archives.nseindia.com/content/indices/ind_nifty50list.csv"

nifty50 = pd.read_csv(url)

print(nifty50.head())
print(f"\nTotal Stocks: {len(nifty50)}")

symbols = nifty50["Symbol"] + ".NS"

print(symbols.tolist())

data = yf.download(
    tickers=symbols.tolist(),
    start="2016-01-01",
    auto_adjust=False,
    actions=False,
    group_by="ticker",
    progress=True,
    threads=True,
)

data.info()

# Basic inspection
print("=" * 80)
print("Shape:", data.shape)
print("\nInfo:")
print(data.info())

print("\nFirst 5 rows:")
print(data.head())

print("\nLast 5 rows:")
print(data.tail())

print("\nColumns:")
print(data.columns)

print("\nMissing values:")
print(data.isna().sum().sort_values(ascending=False).head(20))

print("\nDate Range:")
print(data.index.min(), "->", data.index.max())
```

```python
import yfinance as yf

hourly = yf.download(
    tickers=symbols.tolist(),
    period="730d",      # Maximum practical period for 1h
    interval="1h",
    group_by="ticker",
    auto_adjust=False,
    progress=True,
    threads=True,
)
```

```python
hourly.info()

# Basic inspection
print("=" * 80)
print("Shape:", hourly.shape)
print("\nInfo:")
print(hourly.info())

print("\nFirst 5 rows:")
print(hourly.head())

print("\nLast 5 rows:")
print(hourly.tail())

print("\nColumns:")
print(hourly.columns)

print("\nMissing values:")
print(hourly.isna().sum().sort_values(ascending=False).head(20))

print("\nDate Range:")
print(hourly.index.min(), "->", hourly.index.max())
```

```python
from pathlib import Path
import pandas as pd

# ============================================================
# CONFIG
# ============================================================

datasets = {
    "daily": data,      # Your daily dataframe
    "hourly": hourly     # Your hourly dataframe
}

SAVE_DIR = Path("data/raw/equities")
SAVE_DIR.mkdir(parents=True, exist_ok=True)

# ============================================================
# INSPECTION FUNCTION
# ============================================================

def inspect_and_save(df, name):

    print("\n" + "="*90)
    print(f"{name.upper()} DATASET")
    print("="*90)

    print(f"\nShape           : {df.shape}")
    print(f"Date Range      : {df.index.min()} --> {df.index.max()}")

    mem = df.memory_usage(deep=True).sum()/1024**2
    print(f"Memory Usage    : {mem:.2f} MB")

    print(f"\nRows            : {len(df):,}")
    print(f"Tickers         : {len(df.columns.levels[0])}")
    print(f"Features        : {list(df.columns.levels[1])}")

    print("\nDtypes")
    print(df.dtypes.value_counts())

    # --------------------------------------------------------

    print("\nTop 10 Missing Columns")

    missing_cols = (
        df.isna()
          .sum()
          .sort_values(ascending=False)
    )

    print(missing_cols.head(10))

    # --------------------------------------------------------

    print("\nMissing Values Per Ticker")

    ticker_missing = (
            df.isna()
            .T
            .groupby(level=0)
            .sum()
            .sum(axis=1)
            .sort_values(ascending=False)
        )

    print(ticker_missing)

    # --------------------------------------------------------

    bad_tickers = []

    for ticker in df.columns.levels[0]:

        if df[ticker].isna().all().all():
            bad_tickers.append(ticker)

    print("\n100% Empty Tickers")
    print(bad_tickers if bad_tickers else "None")

    # --------------------------------------------------------

    cleaned = df.copy()

    if bad_tickers:
        cleaned = cleaned.drop(columns=bad_tickers, level=0)

    print("\nAfter Cleaning")

    print(f"Shape      : {cleaned.shape}")
    print(f"Tickers    : {len(cleaned.columns.levels[0])}")

    # --------------------------------------------------------

    print("\nChecking remaining fully-empty tickers...")

    still_bad = []
    tickers = cleaned.columns.get_level_values(0).unique()
    for ticker in tickers:
        if cleaned[ticker].isna().all().all():
            still_bad.append(ticker)

    print(still_bad if still_bad else "PASS")

    # --------------------------------------------------------

    print("\nSample")

    print(cleaned.head())

    # --------------------------------------------------------

    outfile = SAVE_DIR / name /f"nifty50_{name}.parquet"

    cleaned.to_parquet(
        outfile,
        engine="pyarrow",
        compression="snappy"
    )

    print(f"\nSaved -> {outfile}")

    return cleaned


# ============================================================
# RUN
# ============================================================

daily = inspect_and_save(data, "daily")
hourly = inspect_and_save(hourly, "hourly")
```

# NIFTY 50 / Bank Nifty / Midcap indices


```python
benchmark_indices = {
    "NIFTY_50": "^NSEI",
    "BANK_NIFTY": "^NSEBANK",
    "FINNIFTY": "^CNXFIN",
    "NIFTY_100": "^CNX100",
    "NIFTY_200": "^CNX200",
    "NIFTY_500": "^CRSLDX",
    "NIFTY_NEXT_50": "^NSMIDCP",
}
```

```python
import yfinance as yf

indices = yf.download(
    tickers=list(benchmark_indices.values()),
    start="2016-01-01",
    auto_adjust=False,
    group_by="ticker",
    progress=True,
    threads=True,
)

reverse = {v: k for k, v in benchmark_indices.items()}

indices = indices.rename(
    columns=reverse,
    level=0
)

indices = indices.drop(
    columns=["Adj Close", "Volume"],
    level=1,
    errors="ignore",
)

failed = []

for ticker in indices.columns.get_level_values(0).unique():
    if indices[ticker].isna().all().all():
        failed.append(ticker)

print(failed)

indices = indices.drop(columns=failed, level=0)
indices.columns = indices.columns.remove_unused_levels()
```

```python
from pathlib import Path

save_dir = Path("data/raw/market_indices")
save_dir.mkdir(parents=True, exist_ok=True)

indices.to_parquet(
    save_dir / "benchmark_indices.parquet",
    compression="snappy",
)
```

# India GSec \& US Treasury yields (2Y, 10Y, 30Y)

```python
us_treasury = {
    "US_2Y": "DGS2",
    "US_10Y": "DGS10",
    "US_30Y": "DGS30",
}
```

```python
import os
from fredapi import Fred
import pandas as pd
from pathlib import Path
from dotenv import load_dotenv

load_dotenv()

FRED_API_KEY = os.getenv("FRED_API_KEY")

fred = Fred(api_key=FRED_API_KEY)

us_yields = pd.DataFrame({
    name: fred.get_series(series)
    for name, series in us_treasury.items()
})

us_yields.index.name = "Date"

print(us_yields.info())
print(us_yields.head())

save_dir = Path("data/raw/rates")
save_dir.mkdir(parents=True, exist_ok=True)

us_yields.to_parquet(
    save_dir / "us_treasury_yields.parquet",
    compression="snappy",
)
```

# FRED: CPI, yield spreads, credit spreads, industrial production, unemployment, Fed balance sheet

```python
FRED_SERIES = {

    # =====================================================
    # Inflation
    # =====================================================
    "CPI": "CPIAUCSL",
    "CORE_CPI": "CPILFESL",

    # =====================================================
    # Interest Rates
    # =====================================================
    "FED_FUNDS": "FEDFUNDS",
    "SOFR": "SOFR",

    # =====================================================
    # Treasury Yields
    # =====================================================
    "US2Y": "DGS2",
    "US5Y": "DGS5",
    "US10Y": "DGS10",
    "US30Y": "DGS30",

    # =====================================================
    # Yield Curve
    # =====================================================
    "10Y_MINUS_2Y": "T10Y2Y",
    "10Y_MINUS_3M": "T10Y3M",

    # =====================================================
    # Credit Spreads
    # =====================================================
    "BAA_10Y": "BAA10YM",
    "AAA_10Y": "AAA10YM",

    # =====================================================
    # Labor Market
    # =====================================================
    "UNEMPLOYMENT": "UNRATE",

    # =====================================================
    # Industrial Activity
    # =====================================================
    "INDUSTRIAL_PRODUCTION": "INDPRO",

    # =====================================================
    # Consumer Spending
    # =====================================================
    "RETAIL_SALES": "RSAFS",

    # =====================================================
    # Housing
    # =====================================================
    "HOUSING_STARTS": "HOUST",

    # =====================================================
    # Liquidity
    # =====================================================
    "FED_BALANCE_SHEET": "WALCL",

    # =====================================================
    # Dollar
    # =====================================================
    "BROAD_DOLLAR": "DTWEXBGS",

    # =====================================================
    # Recession Indicator
    # =====================================================
    "US_RECESSION": "USREC"
}
```

```python
import os
from fredapi import Fred
import pandas as pd
from pathlib import Path

load_dotenv()
API_KEY = os.getenv("FRED_API_KEY")

fred = Fred(api_key=API_KEY)

macro = pd.DataFrame()

failed = []

for name, series in FRED_SERIES.items():

    try:

        print(f"Downloading {name}")

        macro[name] = fred.get_series(series)

    except Exception as e:

        print(name, e)

        failed.append(name)

macro.index.name = "Date"

print("\nShape:", macro.shape)

print("\nFailed:", failed)

print(macro.info())
```
